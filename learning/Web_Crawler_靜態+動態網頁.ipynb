```python
# lxml 會將xml、html文件轉成元素樹處理
import requests
from lxml import html
from lxml import cssselect
```

```python
res = requests.get('https://www.ptt.cc/bbs/NBA/index.html')
```

```python
res.text
```

```python
# 把 res.text 轉成 tree 解析(一個個節點)
# 可用 css select、xpath 定位
# 使用 html 模組解析它，並將結果保存到 tree 中
tree = html.fromstring(res.text)
```

```python
# tree現在包含了整個 HTML文件到一個優雅的樹結構中，可以使用兩種方法訪問：XPath以及CSS選擇器
tree
```

```python
tree.getchildren()
```

```python
type(tree)
```

```python
tree.xpath('//*[@id="main-container"]/div[2]/div[2]/div[2]/a')
```

```python
for t in tree.xpath('//*[@id="main-container"]/div[2]/div[2]/div[2]/a'):
    print (t.text)
```

```python
# 以 XPath 訪問
for t in tree.xpath('//*[@id="main-container"]/div[2]/div/div/a'):
    print (t.text)
```

```python
# 從套件選取
for t in tree.xpath("//div[@class='title']/a"):
    print (t.text)
```

```python
# 以 CSS Selector 訪問
for t in tree.cssselect(".title a"):
    print (t.text)
```

```python
# #main-container > div.r-list-container.action-bar-margin.bbs-screen > div:nth-child(2) > div.title > a
for t in tree.cssselect('div.title > a'):
    print (t.text)
```
