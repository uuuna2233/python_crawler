```python
movie_names = ["The Devil Wears Prada", "ironman"]

movie_list = []
for i in movie_names:
    url = "http://www.omdbapi.com/?apikey=470ddfd7&t={}".format(i)
    res = requests.get(url)
    movie_list.append(res.json())
movie_list
```

```python
len(movie_list)
```

```python
movies = []
for name in range(len(movie_names)):
    res = requests.get("http://www.omdbapi.com/?apikey=470ddfd7&t={}".format(movie_names[name]))
    movies.append(res.json())
movies
```

```python
enumerate(movie_names)
```

```python
movies = []
for n, name in enumerate(movie_names):
    res = requests.get("http://www.omdbapi.com/?apikey=470ddfd7&t={}".format(name))
    movies.append(res.json())
    
    print(n)
    print(name)
    print("*********")
movies
```

```python
titles = []
years = []
directors = []

for i in range(len(movies)):
    titles.append(movies[i]["Title"])
    years.append(movies[i]["Year"])
    directors.append(movies[i]["Director"])    
    
print(titles)
print(years)  
print(directors)
```

```python
movie3 = []

for i in range(len(movies)):
    moviedic = {"Title":titles[i], "Year":years[i], "Director":directors[i]}
    movie3.append(moviedic)

movie3
```

```python
movie2 = []

for i in range(len(movies)):
    moviedic = {"Title":movies[i]["Title"], "Year":movies[i]["Year"], "Director":movies[i]["Director"]}
    movie2.append(moviedic)

    
movie2
```

```python
movies4 = []
for i, t in enumerate(titles):
    print(i)
    print(t)
    d = {}
    d['Title'] = t
    d['Year'] = years[i]
    d['Director'] = directors[i]
    print(d)
    movies4.append(d)

movies4
```
