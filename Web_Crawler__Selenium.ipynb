```python
import requests
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time, random
from fake_useragent import UserAgent
import pymysql

import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import json
```

```python
keyword = input('請輸入讓心和錢包都不委屈的超夢幻職位：')
```

```python
## Fake User

ua = UserAgent()
options = Options() 
options.add_argument("user-agent=" + ua.chrome)
```

```python
## Chrome Driver

chrome_driver_path = "D:/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path, options=options)

driver.implicitly_wait(10)   # Declared globally, waits maximum 10 seconds for element's presence.

## Search Conditions

my_params = {#'ro':'1',  # Full-time job
             #'kwop':'1',  # Search only by job title 
             'keyword':keyword,
             'isnew':'30',  # Updated within a month 
             'sctp':'M', 'scmin':'40000',  # 40,000 up salary per Month
             'scneg':'1',  # Including negotiable salary
             'jobexp':'1,3',  # 3 years or less of working experience
             'order':'12',  # Sort by relevance
             'mode':'l'}  # List mode

# requests.Response().url, returns the URL of the response
url = requests.get('https://www.104.com.tw/jobs/search/?' , my_params).url
print(url)
driver.get(url)
```

```python
## Scroll to the bottom of the page

Scroll_Pause_Time = 1

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    # Wait to load page
    time.sleep(Scroll_Pause_Time)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
## How many pages we got

# soup.find(tags, attributes, recursive, string, keywords)
soup = bs(driver.page_source, 'lxml')
all_page = int(soup.find('select', 'page-select').option.text.strip(' 頁').split(' / ')[1])
print('搜尋結果共有 ', str(all_page), ' 頁')
```

```python
#@title
## Over page 15 --> Need Manual Loading

k = 16
last_exception, error_counter = None, 0

while 15 < k <= all_page:
    
    try:
        driver.find_elements_by_class_name("js-more-page")[-1].click() 
        # driver.find_element_by_xpath("//*[contains(text(),'手動載入')]").click()
        
        print('Click手動載入第 ', str(k), ' 頁')
        k = k + 1
        if k > all_page:
            print('No more Job')
        time.sleep(random.uniform(2, 3))
        
    # track last error's type, and count it while occurred
    # counter resets if an error of a different type appeared
    # If the counter hits 3, then exit accordingly
    except Exception as e: 
        if type(e) is not last_exception:
            last_exception, error_counter = type(e), 0        
        else:
            error_counter += 1
            if error_counter == 3:
                break
```

```python
## How many jobs we got

# soup.findAll(tags, attributes, recursive, string, limit, keywords)
soup = bs(driver.page_source, 'lxml')
jobs_link = soup.findAll('a', 'js-job-link')

print('搜尋結果共有 ', str(len(jobs_link)), ' 筆')
print(jobs_link)  # Check result format
```
